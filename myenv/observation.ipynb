```python
import pandas as pd
import glob
file_paths = glob.glob("./Datasets/*.json")
dfs = [pd.read_json(file_path) for file_path in file_paths]
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
entries=pd.json_normalize(combined_df.entry)
```

```python
observation_data = entries[entries["resource.resourceType"]=="Observation"]
Diagnostic_data=pd.read_csv("Diagnostic.csv")
```

```python
observation_data.reset_index(drop=True,inplace=True)
```

```python
Diagnostic_data.columns
```

```python
# Assuming 'Diagnostic_data' is your DataFrame
df_diagnostic_data = Diagnostic_data[['Diagnostic_Report_id', 'result_reference']].copy()

# Remove 'urn:uuid:' prefix from 'result_reference' column
df_diagnostic_data.loc[:, 'result_reference'] = df_diagnostic_data['result_reference'].str.replace('urn:uuid:', '')

df_diagnostic_data
```

```python
merged_data = pd.merge(
    df_diagnostic_data,
    observation_data,
    left_on="result_reference", 
    right_on="resource.id",  
    how="right" , 
)
merged_data
```

```python
Observation_type_normalized=pd.json_normalize(merged_data['resource.code.coding'].explode())['display']
Observation_type_normalized
```

```python
# Step 6: Extract the Necessary Fields and Create the Final DataFrame
Observation_output = pd.DataFrame({
    "Observation_id":merged_data['resource.id'],
    'Status': merged_data['resource.status'],
    'patient_id': merged_data['resource.subject.reference'].str.replace('urn:uuid:', ''),
    'Encounter_id':merged_data['resource.encounter.reference'].str.replace("urn:uuid:",''),
    'Observation_Date_Time':merged_data['resource.effectiveDateTime'],
    'Observation_Type':Observation_type_normalized,
    'Observation_value':merged_data['resource.valueQuantity.value'].round(2).astype(str) + merged_data['resource.valueQuantity.unit'],
    'Diagnostic_Report.id':merged_data["Diagnostic_Report_id"],
})
Observation_output
```
