```python
import pandas as pd
import glob
file_paths = glob.glob("./Datasets/*.json")
dfs = [pd.read_json(file_path) for file_path in file_paths]
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
entries=pd.json_normalize(combined_df.entry)
```

```python
CarePlan_data = entries[entries["resource.resourceType"]=="CarePlan"]
CarePlan_data.reset_index(drop=True,inplace=True)
# Encounter_data= entries[entries["resource.resourceType"]=="Encounter"]
```

```python
CarePlan_normalized=pd.json_normalize(CarePlan_data['resource.category'],[0,'coding'])['code']
CarePlan_name=pd.json_normalize(CarePlan_data['resource.category'],[0,'coding'])['display']
```

```python
# Step 6: Extract the Necessary Fields and Create the Final DataFrame
CarePlan_data_output = pd.DataFrame({
    'CarePlan_code':CarePlan_normalized,
    "Status":CarePlan_data['resource.status'],
    "CarePlan_Name":CarePlan_name,
    'patient_id': CarePlan_data['resource.subject.reference'].str.replace('urn:uuid:', ''),
    'Encounter_id':CarePlan_data['resource.context.reference'].str.replace("urn:uuid:",''),
    "CarePlan_Start":CarePlan_data['resource.period.start'],
    "CarePlan_End":CarePlan_data['resource.period.end'],
})
CarePlan_data_output
```

```python
CarePlan_data_output.to_csv("CarePlan.csv",index=False)
CarePlan_data_output.to_excel("CarePlan.xlsx",index=False)

```
