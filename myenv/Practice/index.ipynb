```python
# Extract relevant patient details
patients = []
for entry in data['entry']:
    patient = {}
    patient['name'] = ' '.join(entry['resource']['name'][0]['given']) + ' ' + entry['resource']['name'][0]['family']
    patient['gender'] = entry['resource']['gender']
    patient['birth_date'] = entry['resource']['birthDate']
    patient['marital_status'] = entry['resource']['maritalStatus']['coding'][0]['code']
    patient['city'] = entry['resource']['address'][0]['city']
    patient['state'] = entry['resource']['address'][0]['state']
    patients.append(patient)
```

```python
print(patients)
```

```python
# Convert to DataFrame
df = pd.DataFrame(patients)
print(df)
```

```python
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.show()
```
