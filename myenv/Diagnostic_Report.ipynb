```python
import pandas as pd
import glob
file_paths = glob.glob("./Datasets/*.json")
dfs = [pd.read_json(file_path) for file_path in file_paths]
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
entries=pd.json_normalize(combined_df.entry)
```

```python
Diagnostic_Report_data = entries[entries["resource.resourceType"]=="DiagnosticReport"]
Encounter_data= entries[entries["resource.resourceType"]=="Encounter"]
```

```python
Diagnostic_Report_data.columns
```

```python
encourage_data=Encounter_data[['fullUrl','resource.patient.reference']]
```

```python
merged_data = pd.merge(
    encourage_data,
    Diagnostic_Report_data,
    left_on="fullUrl", 
    right_on="resource.encounter.reference",  
    how="right" , 
)
merged_data
```

```python
for idx, row in merged_data["resource.result"].items():
    print(f"Row {idx}: {row}")
```

```python
normalized_merge_data_diagnostic_name=pd.json_normalize(merged_data['resource.code.coding'].explode())['display']
```

```python
normalized_merge_data_Performer=pd.json_normalize(merged_data['resource.performer'].explode())['display']
```

```python
# Step 6: Extract the Necessary Fields and Create the Final DataFrame
procedure_output = pd.DataFrame({
    "Diagnostic_Report_id":merged_data['resource.id'],
    'Status': merged_data['resource.status'],
    "Diagnostic_Name":normalized_merge_data_diagnostic_name,
    'patient_id': merged_data['resource.patient.reference_x'].str.replace('urn:uuid:', ''),
    'Encounter_id':merged_data['resource.encounter.reference'].str.replace("urn:uuid:",''),
    "Diagnostic_Date_Time": merged_data['resource.effectiveDateTime'],
    "Performer":normalized_merge_data_Performer,
    "result":merged_data['resource.result'],
    # "result_reference":df1,
    # "result_display":df2,
})
procedure_output[:90]
```

```python
procedure_output=procedure_output.explode("result")
```

```python
diagnostic_result_reference = procedure_output['result'].apply(lambda lst:lst['reference'])
diagnostic_result_display = procedure_output['result'].apply(lambda lst:lst['display'])
procedure_output['result_reference'] = diagnostic_result_reference
procedure_output['result_display'] = diagnostic_result_display
procedure_output.drop(columns=['result'],inplace=True)
```
