```python
import pandas as pd
import glob
file_paths = glob.glob("./Datasets/*.json")
dfs = [pd.read_json(file_path) for file_path in file_paths]
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
entries=pd.json_normalize(combined_df.entry)
```

```python
Condition_data = entries[entries["resource.resourceType"]=="Condition"]
Encounter_data= entries[entries["resource.resourceType"]=="Encounter"]
```

```python
Condition_data.columns
```

```python
encounter_data=Encounter_data[['fullUrl','resource.patient.reference','resource.id']]
encounter_data
```

```python
merged_data = pd.merge(
    encounter_data,
    Condition_data,
    left_on="fullUrl", 
    right_on="resource.context.reference",  
    how="right" , 
    suffixes=['_encounter','_condition']
)
merged_data
```

```python
Condition_name_normalized=pd.json_normalize(merged_data['resource.code.coding'].explode())['display']
```

```python
# Step 6: Extract the Necessary Fields and Create the Final DataFrame
condition_output = pd.DataFrame({
    "Condition_id":merged_data['resource.id_condition'],
    'Cleanical_Status': merged_data['resource.clinicalStatus'],
    "Verification_Status":merged_data['resource.verificationStatus'],
    "condition_Name":Condition_name_normalized,
    'patient_id': merged_data['resource.patient.reference_encounter'].str.replace('urn:uuid:', ''),
    'Encounter_id':merged_data['resource.id_encounter'].str.replace("urn:uuid:",''),
    "Onset_Date_Time":merged_data['resource.onsetDateTime'],
    "Abatement_Date_Time":merged_data['resource.abatementDateTime'],
})
condition_output
```

```python
condition_output.to_csv("condition_data.csv",index=False)
condition_output.to_excel("condition_data.xlsx")
```
